```python
from openai import OpenAI
import pandas as pd
```

```python
client = OpenAI(
    api_key="YOUR_OPENAI_API_KEY"
)
```

```python
tasks = [
    "Complete Python assignment",
    "Prepare for interview",
    "Update resume",
    "Apply for remote jobs",
    "Learn OpenAI API"
]

print(tasks)
```

```python
df = pd.DataFrame(tasks, columns=["Task"])

df
```

```python
results = []

for task in df["Task"]:

    response = client.responses.create(
        model="gpt-5",
        input=f"""
        Analyze this task:

        {task}

        Provide:
        - Priority (High, Medium, Low)
        - Estimated Time
        - Productivity Tip
        """
    )

    results.append(response.output_text)
```

```python
df["AI_Analysis"] = results

df
```

```python
for i in range(len(df)):
    print("=" * 60)
    print("Task:", df.loc[i, "Task"])
    print(df.loc[i, "AI_Analysis"])
```

```python
df.to_csv("tasktrack_ai_results.csv", index=False)

print("File saved successfully!")
```
