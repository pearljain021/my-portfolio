```python
import pandas as pd
```

```python
students = {
    "Name": ["Rahul", "Priya", "Aman", "Sneha", "Rohit"],
    "Maths": [85, 72, 91, 65, 78],
    "Science": [88, 75, 95, 70, 80],
    "English": [82, 68, 90, 72, 76]
}

df = pd.DataFrame(students)

df
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]

df
```

```python
df["Percentage"] = df["Total"] / 3

df
```

```python
def grade(percent):
    
    if percent >= 90:
        return "A+"
    
    elif percent >= 80:
        return "A"
    
    elif percent >= 70:
        return "B"
    
    elif percent >= 60:
        return "C"
    
    else:
        return "F"

df["Grade"] = df["Percentage"].apply(grade)

df
```

```python
top_student = df.loc[df["Percentage"].idxmax()]

print("Top Student:")
print(top_student)
```

```python
print(df)
```

```python
df.to_csv("student_grades.csv", index=False)

print("Results saved successfully!")
```
