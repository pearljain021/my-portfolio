```python
from openai import OpenAI
import pandas as pd
```

```python
client = OpenAI(
    api_key="YOUR_OPENAI_API_KEY"
)
```

```python
product = input("Apple: ")
industry = input("iPhone: ")
```

```python
response = client.responses.create(
    model="gpt-5",
    input=f"""
    Conduct market research for:

    Product: {product}
    Industry: {industry}

    Provide:

    1. Market Overview
    2. Target Audience
    3. Main Competitors
    4. SWOT Analysis
    5. Growth Opportunities
    6. Risks and Challenges
    7. Marketing Recommendations
    """
)

analysis = response.output_text

print(analysis)
```
