```python
from openai import OpenAI

client = OpenAI(
    api_key="YOUR_API_KEY"
)
```

```python
headline = """
NTA says no paper leak complaints received during NEET UG 2026 re-exam conducted across India.
"""
```

```python
response = client.responses.create(
    model="gpt-5",
    input=f"""
Analyze the following news headline.

Headline:
{headline}

Provide:

1. Classification (Likely Real or Likely Fake)
2. Confidence Score (0-100%)
3. Reasoning
4. Credibility Assessment
5. Key Facts Mentioned
6. Final Verdict

Format the answer clearly.
"""
)

print(response.output_text)
```

```python
news_headlines = [
    "NTA says no paper leak complaints received during NEET UG 2026 re-exam",
    "22.7 lakh students appear for NEET re-test amid AI and biometric surveillance",
    "Aliens leaked the NEET 2026 question paper on social media",
    "Government cancels all MBBS admissions after NEET 2026"
]

for headline in news_headlines:

    response = client.responses.create(
        model="gpt-5",
        input=f"""
        Analyze this headline:

        {headline}

        Return:
        - Classification
        - Confidence Score
        - Reasoning
        - Final Verdict
        """
    )

    print("=" * 60)
    print("Headline:", headline)
    print(response.output_text)
```
