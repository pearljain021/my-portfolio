```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reviews = [
    "Amazing product and fast delivery",
    "Poor quality and waste of money",
    "The product is okay",
    "Excellent customer service",
    "Bad experience with delivery",
    "Great value for money",
    "Average product",
    "Fantastic packaging and quality",
    "Terrible customer support",
    "Good product for beginners"
]
```

```python
df = pd.DataFrame(reviews, columns=["Review"])

df
```

```python
positive_words = [
    "amazing",
    "excellent",
    "great",
    "good",
    "fantastic"
]

negative_words = [
    "poor",
    "bad",
    "terrible",
    "waste"
]
```

```python
sentiments = []

for review in df["Review"]:

    review = review.lower()

    if any(word in review for word in positive_words):
        sentiments.append("Positive")

    elif any(word in review for word in negative_words):
        sentiments.append("Negative")

    else:
        sentiments.append("Neutral")
```

```python
df["Sentiment"] = sentiments

df
```

```python
print(df["Sentiment"].value_counts())
```

```python
df["Sentiment"].value_counts().plot(kind="bar")

plt.title("Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

```python
df.to_csv("sentiment_analysis_results.csv", index=False)

print("CSV file saved successfully!")
```
