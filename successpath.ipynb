```python
from openai import OpenAI
```

```python
client = OpenAI(
    api_key="YOUR_OPENAI_API_KEY"
)
```

```python
person = """
Dr. A.P.J. Abdul Kalam was an Indian aerospace scientist and the 11th President of India.
He played a leading role in India's missile and space programs and was known as the Missile Man of India.
He inspired millions of students through his speeches, books, and vision for India's development.
"""
```

```python
response = client.responses.create(
    model="gpt-5",
    input=f"""
    Analyze the following personality:

    {person}

    Provide:

    1. Brief Biography
    2. Major Achievements
    3. Leadership Qualities
    4. Challenges Faced
    5. Key Life Lessons
    6. Advice for Students
    7. Success Score (out of 10)
    """
)

print(response.output_text)
```

```python
quote = "Dream, dream, dream. Dreams transform into thoughts and thoughts result in action."

response = client.responses.create(
    model="gpt-5",
    input=f"""
    Analyze this quote by A.P.J. Abdul Kalam:

    {quote}

    Explain:
    1. Meaning
    2. Importance
    3. How students can apply it
    """
)

print(response.output_text)
```

```python
question = input("Ask a question about A.P.J. Abdul Kalam: ")

response = client.responses.create(
    model="gpt-5",
    input=f"""
    Answer this question about Dr. A.P.J. Abdul Kalam:

    {question}
    """
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-5",
    input="""
    Based on the life of Dr. A.P.J. Abdul Kalam,
    provide:

    - Top 10 Success Lessons
    - Career Advice
    - Habits to Develop
    - Skills Students Should Learn
    """
)

print(response.output_text)
```

```python
with open("abdul_kalam_analysis.txt", "w", encoding="utf-8") as file:
    file.write(response.output_text)

print("Analysis saved successfully!")
```
